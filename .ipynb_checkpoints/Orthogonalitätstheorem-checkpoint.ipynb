# Orthogonalitätstheorem
-------------------

# $$ \sum_{a\in G} \Gamma_{i,j}^p(a)\Gamma_{k,l}^{p'}(a^{-1}) = \frac{g}{d}\delta_{i,l}\delta_{j,k}\delta_{p,p'} $$


--------------
## Beweis:

(1) Wir definieren die Matrix:

$$
\tilde{P} = \sum_{a \in G} \tilde{\Gamma}^p(a)\cdot\tilde{Z}\cdot\tilde{\Gamma}^{p'}(a)^{-1}
$$

(2) Für diese Matrix gilt folgende **Identität**

$$
\tilde{P}\cdot\tilde{\Gamma}^{p'}(b) = \tilde{\Gamma}^{p}(b)\cdot\tilde{P}
$$
denn
$$
\begin{align}
\tilde{P}\cdot\tilde{\Gamma}^{p'}(b) &= \sum_{a \in G} \tilde{\Gamma}^p(a)\cdot\tilde{Z}\cdot\tilde{\Gamma}^{p'}(a)^{-1}\tilde{\Gamma}^{p'}(b) \\
&= \sum_{a \in G} \tilde{\Gamma}^p(a)\cdot\tilde{Z}\cdot\left[\tilde{\Gamma}^{p'}(b)^{-1}\tilde{\Gamma}^{p'}(a)\right]^{-1} \\
&= \sum_{a \in G} \left[\tilde{\Gamma}^{p}(b)\cdot\tilde{\Gamma}^{p}(b)^{-1}\right]\tilde{\Gamma}^p(a)\cdot\tilde{Z}\cdot\left[\tilde{\Gamma}^{p'}(b)^{-1}\cdot\tilde{\Gamma}^{p'}(a)\right]^{-1} \\
&= \sum_{a \in G} \tilde{\Gamma}^{p}(b)\cdot\left[\tilde{\Gamma}^{p}(b)^{-1}\cdot\tilde{\Gamma}^p(a)\right]\cdot\tilde{Z}\cdot\left[\tilde{\Gamma}^{p'}(b)^{-1}\cdot\tilde{\Gamma}^{p'}(a)\right]^{-1} \\
&= \tilde{\Gamma}^{p}(b)\cdot\sum_{a \in G} \left[\tilde{\Gamma}^{p}(b)^{-1}\cdot\tilde{\Gamma}^p(a)\right]\cdot\tilde{Z}\cdot\left[\tilde{\Gamma}^{p'}(b)^{-1}\cdot\tilde{\Gamma}^{p'}(a)\right]^{-1}  \\
&= \tilde{\Gamma}^{p}(b)\cdot\sum_{a \in G} \left[\tilde{\Gamma}^{p}(b^{-1}\cdot a)\right]\cdot\tilde{Z}\cdot\left[\tilde{\Gamma}^{p'}(b^{-1}\cdot a)\right]^{-1} \\
&\stackrel{\text{(Umordnungssatz)}}{=}\tilde{\Gamma}^{p}(b)\cdot\tilde{P} \\
\end{align}
$$

(3) Wir haben Z nicht eingeschränkt. Wir wählen Z nun als Matrix, deren Elemente überall null ist ausser in einem Eintrag, dort ist der Eintrag 1:
$$
(Z)_{m,n} = \delta_{m,j}\delta_{n,i}
$$


(4) Nun brauchen wir die **Schur'schen Lemma**:

Teil a:

Seien $D$ und $D'$ irreduzible Matrixgruppen der Ordnung g, d.h. quadratisch selbsabbildend mit der Dimension d bzw d'. 
Wenn es nun eine Matrix **S** der Dimension dxd' gibt, die folgende Gleichung erfüllt:

$$ \tilde{S}\cdot \tilde{D_i'} = \tilde{D_i}\cdot \tilde{S} \space\space\space\space\space\space\space\space\space\forall i $$

So kann **$\tilde{S}$** nur <br>
(a) $\tilde{S}$ = 0 <br>
(b) $\tilde{S}$ regulär, wobei dann d = d' gilt

Teil b:

Wenn 
$$ \tilde{D_i}\cdot \tilde{S} = \tilde{S} \cdot \tilde{D_i} \space\space\space\space\space\space\space\space\space\forall i $$
gilt, also *kommutieren*.

Dann muss $$ \tilde{S} = \lambda \thinspace \tilde{I} $$ 

(5) Nun betrachten wir zwei Fälle:

**Fall 1:** $\tilde{\Gamma}, \tilde{\Gamma}'$ irreduzibel und inäquivalent

das bedeutet dass die beiden Dimensionen nicht übereinstimmen können.
Schauen wir uns nochmal das Ergebnis aus (2) an, sehen wir:
$$
\tilde{P}\cdot\tilde{\Gamma}^{p'}(b) = \tilde{\Gamma}^{p}(b)\cdot\tilde{P} \\

\tilde{\Gamma}^{p}(b)\cdot\tilde{P} = \tilde{P}\cdot\tilde{\Gamma}^{p'}(b)
$$

Sehen wir, dass nach dem "Schurschen Lemme Teil a" automatisch folgt, dass $\tilde{P} = \tilde{S} = 0$. 

**Fall 2:** p = p' und $\overline{\Gamma^{p}}$ irreduzibel

Dann folgt mit Schurschen Lemma Teil b:

$$ \tilde{P} = \lambda \thinspace \tilde{I} $$

Nun bilden wir auf zwei Arten die Spur:

$$
\begin{align*}
    Sp(\tilde{P}) &= Sp(\lambda \thinspace \tilde{I}) = \lambda \cdot d    \\
                  &= g\cdot Sp(\tilde{Z}) = g\cdot \delta_{i,j} 
\end{align*}
$$

Gleichsetzen ergibt somit:

### $$ \lambda = \frac{g}{d} \delta_{i,j}$$

Da wir nun dieses $\lambda$ bestimmen konnten, können wir es rückeinsetzen:

$$ P_{i,l} = \frac{g}{d} \delta_{i,j} \delta_{i,l}   $$

Dieses Ergebnis können wir mit der ursprünglichen Definition von $\tilde{P}$ gleichsetzen, wobei wir berücksichten wie wir unser $\tilde{Z}$
gewählt haben:

$$ P_{i,l} = \frac{g}{d} \delta_{i,j} \delta_{i,l} = \sum_{a \in G} \Gamma_{i,j}^p(a)\Gamma_{k,l}^{p'}(a^{-1}) = \frac{g}{d}\delta_{i,l}\delta_{j,k}\delta_{p,p'} $$
